```python
import pandas as pd

# transforma CSV em Dataframe

arquivocsv = "FlavioBolsonaro_tweets.csv"

df = pd.read_csv(arquivocsv)


#seleciona o per√≠odo
datainicial = "2018-10-28"
datafinal = "2019-02-20"

df['data'] =  pd.to_datetime(df['data'])  
mask = (df['data'] > datainicial) & (df['data'] <= datafinal)
df = df.loc[mask]


#Cria dataframe apenas com retweets
dfRetweets = df.loc[df['texto'].str.startswith('RT')].copy()

#Cria dataframe apenas com tweets originais
dfOriginais = df.loc[~df.index.isin(dfRetweets.index)].copy()

#DfRetweets recebe uma coluna separando o nome das pessoas retuitadas
dfRetweets['origemRt'] = dfRetweets.texto.str.split(expand=True)[1]
```

```python
#Transforma todos os tweets em texto corrido

texto = ""
for linha in dfOriginais.texto:
    texto += linha + " "
```

```python
from nltk.corpus import stopwords 
from nltk.tokenize import TweetTokenizer
        



example_sent = texto

#Palavras mais comuns da l√≠ngua (a serem apagadas)
stop_words = set(stopwords.words('portuguese')) 

#Tokeniza
word_tokens = TweetTokenizer().tokenize(example_sent)

#remove pontua√ß√£o (qualquer coisa com apenas um caracter)
word_tokens = [word_tokens for word_tokens in word_tokens if len(word_tokens) > 1]

#Deixa todos os tokens em min√∫sculas
word_tokens = [word_tokens.lower() for word_tokens in word_tokens]

#apaga todas as palavras mais comuns

#lista com tokens
tokens_filtrados = [] 

for w in word_tokens: 
	if w not in stop_words: 
		tokens_filtrados.append(w) 

#print(word_tokens) 
print(tokens_filtrados) 
```

```python
import nltk
from gensim.models.phrases import Phrases, Phraser

#Cria bigramas: palavras que normalmente estariam separadas ganham um '_' quando h√° muitas ocorrencias

#Treina o algoritmo com a lista de palavras
treinamento = nltk.corpus.mac_morpho.sents()
bigram = Phrases(treinamento, min_count=2, threshold=40)

textoBigramas =(bigram[tokens_filtrados])
```

```python
textoBigramas
```

```python

# Calcula as ocorr√™ncias
fdist = nltk.FreqDist(textoBigramas)

# 100 palavras mais frequentes
dicionariodepalavras = {}
for word, frequency in fdist.most_common(100):
    dicionariodepalavras.update({word:frequency})
```

```python
#Converte as palavras mais frequentes em um DataFrame e depois em um csv
pd.DataFrame(list(dicionariodepalavras.items()), columns=['palavra', 'ocorrencias']).to_csv('PalavrasFrequentes'+arquivocsv)
```
